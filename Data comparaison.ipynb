## Data comparaison

The bulk of the code can be found in load_dat.py and load_xml.py along with some comments explaining the code. My intention has been to find a balance between readability, robustness and efficiency.

I have opted to use DataCompy when comparing the two sets of data. I find it easy to use, fast and most importantly it produces an result that gives a good overview.




```python
xml_df = load_xml.load_xml("animals.xml")
```

```python
dat_df = load_dat.load_dat("animals.dat")
```

```python
def order_columns(df):
    df = df[['id', 'morph', 'miff', 'xcam', 'classA', 'classB', 'classC']]
    return df
```

```python
dat_df=order_columns(dat_df)
dat_df
```

```python
xml_df=order_columns(xml_df)
xml_df
```

```python
compared = dc.Compare(
xml_df,
dat_df,
join_columns='id', 
abs_tol=0,
rel_tol=0,
df1_name='XML',
df2_name='DAT' 
)
```

```python
print(compared.report())
```
